* ### map() function : 



* is used to map values from two series having one column same. For mapping two series, the last column of the first series should be same as index column of the second series, also the values should be unique.

```python
df['Condition of the House'].value_counts()
```

* By analyzing previously implemented plot of condtion sale based on conditon of the house and sale price mean.

 * we can say Good = Fair == 1, 
 * Okay = Bad == 0
 * Excellent = 3

```python
df['Condition of the House'] = df['Condition of the House'].map({'Good': '1',
                                                                 'Excellent':'3',
                                                                 'Bad' : '0',
                                                                 'Fair': '1',
                                                                 'Okay' : '0'})
df['Condition of the House'].unique()
```

* ### Extracting year in separate column from Date House was Sold variable.

```python
def year(value):
    return value.split()[-1]                     # [-1] means from the right side slice it the first element.

df['year_sold'] = df['Date House was Sold'].map(year)
df['year_sold'].head()
```
