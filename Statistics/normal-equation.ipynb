Normal Equation, an analytical approach used for optimization. It is an alternative for Gradient descent. Normal equation performs minimization without iteration.<br>
Normal equations are equations obtained by setting equal to zero the partial derivatives of the sum of squared errors or cost function; normal equations allow one to estimate the parameters of multiple linear regression.

```python
from __future__ import annotations
import matplotlib.pyplot as plt
from typing import List, Tuple
import numpy as np

plt.style.use('seaborn-whitegrid')

from warnings import filterwarnings
filterwarnings("ignore")
```

```python
def create_data() -> None :
    X = 2 * np.random.rand(100, 1)
    y = 4 + 3 * X + np.random.randn(100, 1)
    return X, y
```

```python
X, y = create_data()
plt.figure(figsize = (12, 7), dpi = 80)
plt.plot(X, y, 'r.')
plt.title("Original data")
plt.show()
```

```python
def best_params(X, y) -> None :
    '''
        Normal equation
        theta_best = (X.T * X)^(-1) * X.T * y
    '''
    X_transpose = X.T
    theta_best = np.linalg.inv(X_transpose.dot(X)).dot(X_transpose).dot(y)
    return theta_best
```

we've not chosen to add some noise, this data would've been a dead straight line, and then we wouldn't need fancy machine learning. Simple math would do. 

```python
X_b = np.c_[np.ones((100, 1)), X] # set bias term to 1 for each sample  
params = best_params(X_b, y)
print(params)
```

```python
# test prediction

test_X = np.array([[0], [2]])
test_X_b = np.c_[np.ones((2, 1)), test_X]

prediction = test_X_b.dot(params)
print(prediction)
```

```python
plt.figure(figsize = (12, 7), dpi = 80)
plt.plot(X, y, 'b.')
plt.plot(test_X, prediction, 'r--')
plt.legend([])
plt.title("best fit line")
plt.show()
```
