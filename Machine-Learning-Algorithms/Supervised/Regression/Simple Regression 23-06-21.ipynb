```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('salary_data.csv')
data.head()
```

### Dividing Independent and Dependent variables from train, test split.

```python
x = data.iloc[:,:-1]
y = data.iloc[:,1:2]
```

# Simple Linear model :

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)

from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
```

```python
linear_reg.fit(x_train, y_train)
```

```python
y_pred = linear_reg.predict(x_test)
```

```python
linear_reg.predict([[2.4]])
```

```python
print('Estimated Salary : ',linear_reg.predict([[1.1]]))
```

## Checking accuracy :

```python
print('Accuray of our model :', linear_reg.score(x_test, y_test)*100)
```
